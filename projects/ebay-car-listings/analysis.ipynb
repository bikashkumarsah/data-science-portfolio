To decide the hightest and lowest acceptable values for registration year. We choose the range 1950 to 2016 as the car can't be registered in the future, because of its listing in the ebay. 1950 is lower acceptable range because some other the seller might want to sell their old vintage car.

```python
autos = autos[autos['registration_year'].between(1900, 2016)]
autos['registration_year'].value_counts(normalize = True).sort_index(ascending = False)
```

We can see that aroud 48000 cars fall in this date range.

```python
 auto = autos['brand'].value_counts(normalize = True)
auto
```

I am choosing to  aggregate the value of top 5 brands.

```python
price_dict = {}
for index in auto.index[:5]:
    brand = autos[autos['brand'] == index]
    mean_price = brand['price'].mean()
    price_dict[index] = int(mean_price)
    
```

```python
print(price_dict)
```

We can see from the value that mercedes benz has the highest value overall compared to other.

```python
milage_dict = {}
for index in auto.index[:5]:
    brand = autos[autos['brand'] == index]
    mean_milage = brand['odometer_km'].mean()
    milage_dict[index] = int(mean_milage)
milage_dict
```

```python
milage = pd.Series(price_dict)
price = pd.Series(milage_dict)
```

```python
df = pd.DataFrame(milage, columns = ['mean_milage'])
df_price = pd.DataFrame(price, columns = ['mean_price'])
df['mean_price'] =  df_price
```
