# Star Wars Survey

```python
import pandas as pd
# Read the data
star_wars = pd.read_csv('../../dataset/star_wars.csv', encoding = 'ISO-8859-1')
```

```python
# Read the first 10 row of star_war dataset
star_wars.head(10)
```

```python
# Listing the columns
star_wars.columns
```

## Converting yes/no column to Boolean

```python
data = {'Yes':True, 'No':False}
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(data)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(data)
```

```python
star_wars.head()
```

## Cleaning and mapping the checkbox columns

```python
import numpy as np
data_clean = {
    "Star Wars: Episode I  The Phantom Menace": True,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(data_clean)
    star_wars[col] = star_wars[col].fillna(False).astype(bool)
```

```python
star_wars.head()
```

```python
star_wars = star_wars.rename(columns = {'Which of the following Star Wars films have you seen? Please select all that apply.':'seen_1',
                            'Unnamed: 4':'seen_2', 'Unnamed: 5':'seen_3',
                            'Unnamed: 6':'seen_4', 'Unnamed: 7':'seen_5',
                            'Unnamed: 8':'seen_6'})
                 
```

```python
star_wars.head()
```

```python
# Converting the columns to numeric
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
star_wars = star_wars.rename(columns = {'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.':'ranking_1',
                                        'Unnamed: 10':'ranking_2', 'Unnamed: 11':'ranking_3',
                                        'Unnamed: 12':'ranking_4', 'Unnamed: 13':'ranking_5',
                                        'Unnamed: 14':'ranking_6'})
star_wars.head()
```

```python
%matplotlib inline
```

```python
# Computing the mean rank of the ranking column
bar_plot = star_wars[star_wars.columns[9:15]].mean(numeric_only = True)
```

```python
bar_plot.plot.bar()
```

## Summary
So far, we've cleaned up the data, renamed several columns, and computed the average ranking of each movie. As we suspected, it looks like the "original" movies are rated much more highly than the newer ones.
