## Core functions

```python
# factorial function
def factorial(n):
    final_sum = 1
    for i in range(n, 0, -1):
        final_sum *= i
    return final_sum

# function to calculate combinations.
def combinations(n, k):
    numerator = factorial(n)
    denominator = factorial(k) * factorial(n-k)
    comb = numerator / denominator
    return comb
                           
```

## One-ticket Probability

Let's understand the rule of 6/49 game.

In the 6/49 lottery, six numbers are drawn from a set of 49 numbers that range from 1 to 49. A player wins the big prize if the six numbers on their tickets match all the six numbers drawn. If a player has a ticket with the numbers {13, 22, 24, 27, 42, 44}, he only wins the big prize if the numbers drawn are {13, 22, 24, 27, 42, 44}. If only one number differs, he doesn't win.

The requirement of our function.

- Inside the app, the user inputs six different numbers from 1 to 49.
- Under the hood, the six numbers will come as a Python list, which will serve as the single
- The engineering team wants the function to print the probability value in a friendly way — in a way that people without any probability training are able to understand.

```python
# function to calculate one_ticket_probability

def one_ticket_probability(array):
    possible_numbers = 49
    total_combinations = combinations(possible_numbers, len(array))
    successful_outcome = 1
    probability_of_success = successful_outcome / total_combinations
    print('''Your chances to win the big prize with the numbers {} are {:.7f}%.
In other words, you have a 1 in {:,} chances to win.'''.format(array,
                    (probability_of_success * 100), int(total_combinations)))
```

```python
# Test the function
one_ticket_probability([1,2,3,4,5,6])
```

## Multi-ticket Probability
We're going to write a function that will allow the users to calculate the chances of winning for any number of different tickets.

The requirement of the function.

- The user will input the number of different tickets they want to play (without inputting the specific combinations they intend to play).

- Our function will see an integer between 1 and 13,983,816 (the maximum number of different tickets).
- The function should print information about the probability of winning the big prize depending on the number of different tickets played.

```python
# function to calculate multi_ticket probability

def multi_ticket_probability(no_of_ticket):
    combination = combinations(49, 6)
    probability_of_success = no_of_ticket /  combination
    print('''Your chances to win the big prize with  {}  tickets are {:.7f}.
In other words, you have a {:,} in 13,983,816 chances to win.'''.format(no_of_ticket, probability_of_success * 100, no_of_ticket))
    
```

```python
list = [1, 10, 100, 10000, 1000000, 6991908, 13983816]
for num in list:
    multi_ticket_probability(num)
    print('----------------------------------------------------------\n')
```

## Less Winning Numbers

We're going to write one more function to allow the users to calculate probabilities for two, three, four, or five winning numbers.

The requirement of the function are:

-  the user inputs six different numbers from 1 to 49.
-  an integer between 2 and 5 that represents the number of winning numbers expected
-  Our function prints information about the probability of having the inputted number of winning numbers.

```python
def probability_less_6(num_of_input):
    num_combination_ticket = combinations(6, num_of_input)
    num_remaining_combination = combinations(43, 6 - num_of_input)
    no_of_successful_outcome = num_combination_ticket * num_remaining_combination
    

    total_combination = combinations(49, 6)
    probability_of_success = no_of_successful_outcome / total_combination
    combinations_simplified = round(total_combination/no_of_successful_outcome)
    
    print('''Chances of having {} winning number in this ticket are {:.6f}%
    In other words you have 1 in {} chances to win the lottery. '''.format(num_of_input, probability_of_success * 100, combinations_simplified))
```

```python
inputs = [2,3,4,5]
for num in inputs:
    probability_less_6(num)
    print('------------------------------------------------------\n')
```
