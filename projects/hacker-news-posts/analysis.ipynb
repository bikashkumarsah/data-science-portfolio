# Analysing Hacker News posts.
The analysis mainly focus of analyzing Ask HN or Show HN. The post total reaction and coments and analyze the influence and time on post like and comment.

```python
from csv import reader
with open('../../dataset/hacker_news.csv') as file:
    file = reader(file)
    hn = list(file)
print(hn[:5])
```

```python
headers = hn[0]
print(headers)
hn = hn[1:]
print(hn[:5])
```

```python
ask_posts = []
show_posts = []
other_posts = []
for row in hn:
    title = row[1].lower()
    if title.startswith('ask hn'):
        ask_posts.append(row)
    elif title.startswith('show hn'):
        show_posts.append(row)
    else:
        other_posts.append(row)
print(len(ask_posts), len(show_posts), len(other_posts))
```

```python
total_ask_comments = 0
for row in ask_posts:
    comment = int(row[4])
    total_ask_comments += comment
avg_ask_comments = total_ask_comments / len(ask_posts)
print(avg_ask_comments)
total_show_comments = 0
for row in show_posts:
    comment = int(row[4])
    total_show_comments += comment
avg_show_comments = total_show_comments / len(show_posts)
print(avg_show_comments)
    
```

After analysing the ask_posts and comment_posts list. We can observe that on average the ask_posts receive 14 comments while the show_posts receive 10 comments. This show that the ask_post have 40% more comment compared to show_posts.

```python
import datetime as dt
result_list = []
counts_by_hour = {}
comments_by_hour = {}
counter = 0
for row in ask_posts:
    created = row[6]
    comment = int(row[4])
    result_list.append([created, comment])
print(result_list[:5])
```

```python
for row in result_list:
    comment = int(row[1])
    date = dt.datetime.strptime(row[0], '%m/%d/%Y %H:%M')
    date = dt.datetime.strftime(date, '%H')
    if date in counts_by_hour:
        counts_by_hour[date] += 1
        comments_by_hour[date] += comment
    else:
        counts_by_hour[date] = 1
        comments_by_hour[date] = comment
print(counts_by_hour)
print(comments_by_hour)
    
```

```python
average_by_hour = []
for comment in comments_by_hour:
    average_by_hour.append([comment, comments_by_hour[comment] / counts_by_hour[comment]])
    
print(average_by_hour)
```

```python
swap_avg_by_hour = []
for row in average_by_hour:
    swap_avg_by_hour.append([row[1], row[0]])
sorted_swap = sorted(swap_avg_by_hour,reverse = True)
print(sorted_swap)

```

```python
print(f'Top 5 Hours for Ask Posts Comments')
for row in sorted_swap[:5]:
    date = dt.datetime.strptime(row[1], '%H')
    date = dt.datetime.strftime(date, '%H:%M')
    print(f'{date}: {row[0]:.2f} average comments per post')
```
